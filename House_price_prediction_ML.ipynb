```
# decrease the skewnwnes of the data
for i in skewed_features:
    dataset[i] = np.log(dataset[i] + 1)
```

```
SalePrice = np.log(x['SalePrice'] + 1)
```

```
# Creating list of all the values which have to be used as  Dummy Variables 
dummy_list=['MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities',
       'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
       'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
       'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation',
       'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
       'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual',
       'Functional', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual',
       'GarageCond', 'PavedDrive', 'PoolQC', 'Fence', 'MiscFeature',
       'SaleType', 'SaleCondition']
```

```
dummy_drop = []
clean_df = dataset
for i in dummy_list:
    dummy_drop += [i + '_' + str(dataset[i].unique()[-1])]

dataset = pd.get_dummies(dataset, columns = dummy_list)
dataset = dataset.drop(dummy_drop, axis = 1)
```

```
# removing -infy values from datafram 
dataset.replace([np.inf, -np.inf], np.nan).dropna(axis=0)
```

```
train_len = len(x)
X_train = dataset[:train_len]
X_test = dataset[train_len:]
y_train = SalePrice

print(X_train.shape)
print(X_test.shape)
print(len(y_train))
```

```
X_train=X_train.drop(['SalePrice'],axis=1)
```

```
X_test.drop(['SalePrice'],axis=1,inplace=True)
```

```
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import make_scorer, r2_score

def test_model(model, X_train=X_train, y_train=y_train):
    cv = KFold(n_splits = 3, shuffle=True, random_state = 45)
    r2 = make_scorer(r2_score)
    r2_val_score = cross_val_score(model, X_train, y_train, cv=cv, scoring = r2)
    score = [r2_val_score.mean()]
    return score
```

```
from sklearn.svm import SVR
svr_reg = SVR(kernel='rbf')
test_model(svr_reg)
```

```
import sklearn.linear_model as linear_model
LR = linear_model.LinearRegression()
test_model(LR)
```

```
rdg = linear_model.Ridge()
test_model(rdg)
```

```
svr_reg.fit(X_train,y_train)
y_pred = np.exp(svr_reg.predict(X_test)).round(2)
```
